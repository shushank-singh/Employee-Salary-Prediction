```python
df = pd.read_csv("../data/processed_data/clean_data.csv")
```

```python
df.head()
```

```python
numericals_col = [col for col in df.columns if df[col].dtype in ["int64",float]]
categoricals_col = [col for col in df.columns if df[col].nunique()>1 and df[col].dtype == "str" and col != "Churn"]
```

### Encoding

```python
encode = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
df_encode = pd.DataFrame(
    encode.fit_transform(df[categoricals_col]),
    columns=encode.get_feature_names_out(categoricals_col),
    index = df.index 
)
```

```python
scale = StandardScaler()
df_scale = pd.DataFrame(
    scale.fit_transform(df[numericals_col]),
    columns = scale.get_feature_names_out(numericals_col),
    index = df.index
)
```

```python
df_scale
```

```python
final_df = pd.concat([df_scale,df_encode],axis=1)
```

```python
final_df.to_csv("../data/processed_data/feature_data.csv", index=False)
```
